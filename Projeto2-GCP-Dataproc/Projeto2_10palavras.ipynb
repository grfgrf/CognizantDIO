```python
import pandas as pd
```

```python
df = pd.read_csv('gs://diobucket1/resultado/part-00000',sep='\n',header=None)
```

```python
palavra_e_count = df[0].str.replace('(', '', regex=False).str.replace(')', '', regex=False).str.rsplit(',', n=1, expand=True)
palavra_e_count.rename(columns = {0: 'palavra', 1: 'qnt'} , inplace = True)
palavra_e_count.head(3)
palavra_e_count.dtypes
```

```python
#to_int garantir ordenação. 
palavra_e_count['qnt'] = palavra_e_count['qnt'].astype('int') 
palavra_e_count.dtypes
```

```python
#fetch 10 rows desc
top10 = palavra_e_count[:10].sort_values('qnt', ascending=False)
top10.head(3)
```

```python
#resultado Series
resultado = top10['palavra'].str.replace("'",'',regex=False)
resultado
```

```python
resultado.to_csv('gs://diobucket1/resultado/resultado.txt',index=False, header=None)
```
